### 1.3. Limpieza de los Datos


Dentro de está sección lo que vamos a controlar son, tanto los valores nulos como los valore atípicos


#### 1.3.1. Valores Nulos

```python
missing_values_count = df.isnull().sum()
```

```python
missing_values_count[0:10]
```

```python
print(np.prod(df.shape))
```

```python
# how many total missing values do we have?
total_cells = np.prod(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

El 15,83% de las celdas contienen valores nulos, por lo que es un porcentaje muy elevado, entonces ahora nos vamos a encargar de corregirlo
